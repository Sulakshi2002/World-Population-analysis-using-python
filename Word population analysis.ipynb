### Visualization

```python
# Draw a histogram with 10 bars showing the distribution of the 2023 population column, 
# and also draw a smooth curve showing the density of the distribution.
#KDE curve (Kernel Density Estimate)It is a smooth curve that shows the shape of your data distribution.
```

```python
sns.histplot(df["2023 population"], bins = 10, kde = True)
plt.title("Distribution of 2023 Population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("continent")["2023 population"].mean()
```

```python
df.groupby("continent")["2023 population"].mean().reset_index()
```

```python
df.groupby("continent")["2023 population"].mean().max()
```

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending = False)
```

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending = False).head(2)
```

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending = False).tail(2) # no ascending,descending,last 2 values
```

```python
top10_countries = df.sort_values(by="2023 population", ascending = False).head(10)
```

```python
top10_countries
```

```python
sns.barplot(x="2023 population", y="country", data = top10_countries)
plt.title("Top 10 countries by population in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
df.rename(columns = {"density (km²)" : "density (km2)"},inplace = True)
```

```python
df.columns
```

```python
sns.scatterplot(x="2023 population", y = "density (km2)", data=df)
plt.title("Population and Density Scatter Plot")
plt.xlabel("2023 Population")
plt.ylabel("Density")
plt.show()
```

```python
df.rename(columns = {"area (km²)" : "area (km2)"},inplace = True)
```

```python
df.groupby("continent")["area (km2)"].sum().sort_values( ascending = False)
```

```python
sns.violinplot(
    x="continent",
    y="density (km2)",
    data=df,
    color="lightgreen"
)
plt.title("Population Density distribution across continents")
plt.xlabel("Continent")
plt.ylabel("Population Density (km²)")
plt.show()
```

```python
#violin plot:
#Distribution shape like violin shape
#Fat/wide areas = many observations
#Thin/narrow areas = fewer observation
```

```python
sns.boxplot(
    x="continent",
    y="density (km2)",
    data=df,
    color="purple",
    showfliers = False #Hide the outliers from the plot.
)
plt.title("Population Density distribution across continents (Without Outliers)")
plt.xlabel("Continent")
plt.ylabel("Population Density (km²)")
plt.show()
```
